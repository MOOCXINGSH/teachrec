```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt

f500 = pd.read_csv("M02_wkreport.csv", encoding = 'utf8',index_col=0)
print(f500.head(10))
print(f500.tail(10))
print(f500.info())

weeksch=np.zeros([5,7])

dr=pd.date_range("2018-01-01 0:0", periods=5,freq="30min")
#print(dr)
strtime=[]
empty=[]
for date in dr:
    strtime.append(str(date)[11:16])
    empty.append("  ")
print(date.dayofweek)

weekdef=["周一","周二","周三","周四","周五","周六","周日"]
df=pd.DataFrame(weeksch,columns=weekdef)

for wk in weekdef:
    df[wk]=empty
#df["时间"]=strtime

enddate=str(datetime.date.today())

dr=pd.date_range("2018-03-01",enddate,freq="D")
line=0
for date in dr:
    strdate=str(date)[0:10]
    dayofweek=date.dayofweek
    print(strdate,dayofweek)
    todaywkrec=f500[f500["Workdate"]==strdate]
    if not todaywkrec.empty:
        teachers=todaywkrec["Teacher"]
        teacherstr=""
        for teacher in teachers:
            teacherstr+=" "+str(teacher)
        teacherstr+="|"
        wkmaps=todaywkrec["WkRecMap"]
        for wkmap in wkmaps:
            students=str(wkmap).split("|")[1]
            teacherstr+=students
        df.loc[line,weekdef[dayofweek]]=teacherstr
        
    else:
        df.loc[line,weekdef[dayofweek]]=strdate[5:10]
    if dayofweek==6:
        line+=1
print(df)
```
